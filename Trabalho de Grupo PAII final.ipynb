# Top 10 influencers com mais subscritores no TikTok 

```python
#aqui cri√°mos uma variavel que compreende os 10 influencers com mais subs do tiktok
top10subs = top.sort_values(by=['Subscriberscount'], ascending=False).head(10)
top10subs
#a fun√ß√£o sort_values organiza por defeito os dados por ordem ascendente, logo recorremos √† fun√ß√£o ascending=false para que fizesse o contr√°rio e termos os dados organizados por ordem de quem tem mais subs
#head(10) mostra as 10 primeiras linhas, neste caso os 10 influencers com mais subscritores
```

# Gr√°fico de Subscritores

```python
import numpy as np #importar os numpy como np
import matplotlib.pyplot as plt #importar da biblioteca dados de gr√°ficos


#dados necess√°rios para o gr√°fico
conta = top10subs ['Account'].tolist() #transformar os valores da coluna Account presentes numa lista numa √∫nica vari√°vel
subs = top10subs ['Subscriberscount'].tolist() #transformar os valores da coluna Subscriberscount presentes numa lista numa √∫nica vari√°vel

#defini√ß√µes do gr√°fico
plt.bar(conta, subs, color="blue") #cria√ß√£o de um gr√°fico de barras azul com os dados das colunas Subscriberscount
plt.gcf().set_size_inches(20, 6) #fun√ß√£o que permite trabalhar as dimens√µes do gr√°fico

#estiliza√ß√£o do gr√°fico
plt.xlabel("Contas") #t√≠tulo do eixo do x (horizontal)
plt.title("N√∫mero de subscritores") #t√≠tulo do gr√°fico
plt.xticks(conta) #valores do eixo do x (todos os valores da coluna Account)
plt.yticks([500000,50000000,100000000,150000000]) #escala de valores do eixo y (vertical)
plt.show() #invocar o gr√°fico de barras
```

# Compara√ß√£o entre as views e subs do top 10 influencers com mais subscritores no tiktok.

```python
import numpy as np #importar os numpy como np
import matplotlib.pyplot as plt #importar da biblioteca dados de gr√°ficos


#dados para o gr√°fico
conta = top10subs ['Account'].tolist() #transformar os valores da coluna Account numa lista numa √∫nica vari√°vel
subs = top10subs ['Subscriberscount'].tolist() #transformar os valores da coluna Subscriberscount numa lista numa √∫nica vari√°vel
conta = top10subs ['Account'].tolist() #transformar os valores da coluna Account em uma lista numa √∫nica vari√°vel
views = top10subs ['Viewsavg'].tolist() #transformar os valores da coluna Viewsavg em uma lista numa √∫nica vari√°vel

#defini√ß√µes do gr√°fico
plt.bar(conta, subs, color="blue", label="subscritores") #criar um gr√°fico de barras azul com os dados das colunas Subscriberscount
plt.gcf().set_size_inches(20, 6) #fun√ß√£o para trabalhar as dimens√µes do gr√°fico
plt.bar(conta, views, color="red", label="visualiza√ß√µes") #criar um gr√°fico de barras azul com os dados das colunas Viewsavg
plt.gcf().set_size_inches(20, 6) #fun√ß√£o para trabalhar as dimens√µes do gr√°fico

#estiliza√ß√£o do gr√°fico
plt.xlabel("Contas") #t√≠tulo do eixo do x (horizontal)
plt.title("N√∫mero de subscritores perante visualiza√ß√µes") #t√≠tulo do gr√°fico
plt.xticks(conta) #valores do eixo do x (todos os valores da coluna Account)
plt.yticks([500000,50000000,100000000,150000000]) #escala para os valores do eixo y (vertical)
plt.legend() #mostrar legenda do gr√°fico
plt.show() #mostrar o gr√°fico
```

# Gr√°fico compara√ß√£o entre likes e coment√°rios dos influencers com a menor m√©dia de likes 

```python
import numpy as np #importar os numpy como np
import matplotlib.pyplot as plt #importar da biblioteca dados de gr√°ficos


#dados para o gr√°fico
conta = top10likes ['Account'].tolist() #transformar os valores da coluna Account numa lista numa √∫nica vari√°vel
likes = top10likes ['Likesavg'].tolist() #transformar os valores da coluna Likesavg numa lista numa √∫nica vari√°vel
conta = top10likes ['Account'].tolist() #transformar os valores da coluna Account numa lista numa √∫nica vari√°vel
comments = top10likes ['Commentsavg'].tolist() #transformar os valores da coluna Commentavg numa lista numa √∫nica vari√°vel

#defini√ß√µes do gr√°fico
plt.bar(conta, likes, color="blue", label="Likes") #criar um gr√°fico de barras azul com os dados das colunas Likesavg
plt.gcf().set_size_inches(20, 8) #fun√ß√£o para trabalhar as dimens√µes do gr√°fico
plt.bar(conta, comments, color="red", label="Coment√°rios") #criar um gr√°fico de barras azul com os dados das colunas Commentavg
plt.gcf().set_size_inches(20, 8) #fun√ß√£o para trabalhar as dimens√µes do gr√°fico

#estiliza√ß√£o do gr√°fico
plt.xlabel("Contas") #t√≠tulo do eixo do x (horizontal)
plt.title("N√∫mero de subscritores perante visualiza√ß√µes") #t√≠tulo do gr√°fico
plt.xticks(conta) #valores do eixo do x (todos os valores da coluna Account)
plt.yticks([10000,30000,50000,70000,90000]) #escala para os valores do eixo y (vertical)
plt.legend() #legenda do gr√°fico
plt.show() #mostrar o gr√°fico
```

# Qual √© a m√©dia de partilhas do top 1000 de influencers no TikTok?

```python
media_partilhas = top["Sharesavg"].mean() #m√©dia de todos os valores da coluna Sharesavg

media_partilhas=round(media_partilhas,2) #arredondar o resultado para 2 casas decimais

print("No top 1000 do TikTok, existe em m√©dia ",media_partilhas , "de partilhas.")
```

# Qual √© a m√©dia de coment√°rios do top 1000 de influencers no TikTok?

```python
media_comentarios = top["Commentsavg"].mean() #m√©dia de todos os valores da coluna Commentsavg

media_comentarios=round(media_comentarios,2) #arredondar o resultado para 2 casas decimais

print("No top 1000 do TikTok, existe em m√©dia ",media_comentarios , "de coment√°rios.")
```

# Qual √© a m√©dia de likes do top 1000 de influencers no TikTok?

```python
media_likes = top["Likesavg"].mean() #m√©dia de todos os valores da coluna Likesavg

media_likes=round(media_likes,2) #arredondar o resultado para 2 casas decimais

print("No top 1000 do TikTok, existe em m√©dia ",media_likes , "de likes.")
```

# Qual √© a m√©dia de Views do top 1000 de influencers no TikTok?

```python
media_views = top["Viewsavg"].mean() #m√©dia de todos os valores da coluna Viewsavg

media_views=round(media_views,2) #arredondar o resultado para 2 casas decimais

print("No top 1000 do TikTok, existe em m√©dia ",media_views , "de views.")
```

# Qual √© a m√©dia de Subs do top 1000 de influencers no TikTok?

```python
media_subs = top["Subscriberscount"].mean() #m√©dia de todos os valores da coluna Subscriberscount

media_subs = round(media_subs,2) #arredondar o resultado para 2 casas decimais

print("No top 1000 do TikTok, existe em m√©dia ",media_subs, "de subscritores.")
```

```python

#dados do gr√°fico
labels = ['Subscritores', 'Visualiza√ß√µes', 'Likes', 'Coment√°rios', 'Partilhas'] #dar nome as legendas do gr√°fico
Medias = [media_subs, media_views, media_likes , media_comentarios, media_partilhas] #definir quais os dados a serem colocados 



plt.axis("equal") #define uma escala igual

plt.pie(Medias, labels=labels, autopct='%1.1f%%',shadow=True, radius =1.5)#fazer um gr√°fico circular com os 
#dados das M√©dias calculadas.foram postas labels - subscritores, visualiza√ß√µes, likes, coment√°rios e partilhas.
#Autopct serve para atribuir uma legenda a cada uma das "fatias" do gr√°fico
#o seu respetivo n√∫mero.shadow cria uma sombra no gr√°fico quando igualdo a true 
#Radius √© o raio do gr√°fico que inicialmente √© atribuido automaticamente o valor de 1.

plt.legend(loc='lower right', bbox_to_anchor=(1.8,-0.1))#plt.legend coloca a legenda do gr√°fico no s√≠tio escolhido, neste caso canto inferior direito. 
#loc='upper left' = localiza√ß√£o mias precisa da legenda - canto superior esquerdo
plt.title('Influencers', pad= 70)#plt.title = (t√≠tulo do gr√°fico que neste caso foi subido quanto ao eixo superior em 70 pontos


plt.show() #mostrar o gr√°fico de barras
```
